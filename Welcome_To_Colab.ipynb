```
#Melani Indriyani (251011700794)
# SOAL 1: KONVERSI HARI
print("Konversi Hari")
hari = int(input("Masukkan Total Hari: "))

tahun = hari // 365
sisa_hari = hari % 365

bulan = sisa_hari // 30
hari_akhir = sisa_hari % 30

print("Hasil Konversi:")
print("Tahun:", tahun)
print("Bulan:", bulan)
print("Hari:", hari_akhir)

#Melani Indriyani (251011700794)
# SOAL 2: FIZZBUZZ
print("FizzBuzz")
x = int(input("Masukkan angka maksimal (1-100): "))

if x > 100:
    print("Maksimal hanya sampai 100")
elif x < 1:
    print("Masukkan angka minimal 1")
else:
    for i in range(1, x + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)
```
